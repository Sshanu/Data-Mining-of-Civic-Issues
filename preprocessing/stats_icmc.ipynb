```python
import glob
import cv2 as cv
import json
import numpy as np
from collections import Counter
import operator
```

```python
reports = []
for file in glob.glob("../data/icmc/complaints_*.json"):
    for line in open(file, "r"):
        reports.append(json.loads(line))
```

```python
len(reports)
```

```python
reports[0]
```

```python
_categories = []

for report in reports:
    _categories.append(report['category'])
```

```python
categories = set(_categories)
```

```python
count_categories = Counter(_categories).most_common()
```

```python
count_categories
```

```python
_subCategories = []

for report in reports:
    _subCategories.append(report['subCategory'])
```

```python
subCategories = set(_subCategories)
```

```python
count_subCategory = Counter(_subCategories).most_common()
```

```python
count_subCategory
```

```python
num_image = dict((s, 0) for s in subCategories)
```

```python
for rep in reports:
    num_image[rep['subCategory']] += len(rep['image_url'])
```

```python
num_image = sorted(num_image.items(), key=operator.itemgetter(1), reverse=True)
```

```python
num_image
```

```python
_descriptions = []
for report in reports:
    _descriptions.append(report['description'])
```

```python
_words = []

for des in _descriptions:
    for w in des.split():
        _words.append(w.lower())
```

```python
count_words = Counter(_words).most_common()
```

```python
count_word_dict = dict((w, c) for w, c in count_words)
```

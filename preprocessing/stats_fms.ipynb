```python
import glob
import cv2 as cv
import json
import numpy as np
from collections import Counter
import operator
```

```python
reports = []
for file in glob.glob("../data/fms/complaints_*.json"):
    print(file)
    for line in open(file, "r"):
        try:
            reports.append(json.loads(line))
        except:
            print('error')
```

```python
len(reports)
```

```python
_categories = []

for report in reports:
    _categories.append(report['category'])
```

```python
categories = set(_categories)
```

```python
count_categories = Counter(_categories).most_common()
```

```python
def stats(category):
    num_images = 0
    num_reports = 0
    num_pairs = 0
    num_descriptions = 0
    for rep in reports:
        if(rep['category'] == category):
            num_reports += 1
            if(rep['images'] != None) and (len(rep['images']))>0:
                num_images += len(rep['images'])
                if (len(rep['descroption'])>0 and rep['descroption'][0]!=None):
                    num_pairs += 1
            if(len(rep['descroption'])>0 and rep['descroption'][0]!=None):
                num_descriptions +=1
            
    return num_reports, num_descriptions, num_images, num_pairs
```
